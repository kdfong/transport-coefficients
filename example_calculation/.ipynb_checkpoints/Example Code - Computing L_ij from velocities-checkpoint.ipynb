# Example calculation of transport coefficients L$^{ij}$

This code uses output files from LAMMPS molecular dynamics simulations to compute transport coefficients $L^{ij}$ in electrolyte solutions, where $L^{ij} = \frac{V}{3 k_B T} \int_0^\infty dt \big<c_i (\mathbf{v}_{i}(t)-\mathbf{v}(t)) \cdot c_i (\mathbf{v}_{j}(t)-\mathbf{v}(0))\big>$.

Here, $i$ and $j$ are ionic species in solution, $c_i = \frac{N_i}{V}$ is the concentration of species $i$, $\mathbf{v}_i$ is the velocity of species $i$, and $\mathbf{v}$ is the mass-averaged velocity of all species in the system.

For the purpose of this computation, we assume $\mathbf{v} = 0$, i.e. there is negligible center of mass velocity of the simulation box. Note that while $\mathbf{v}$ will impact $L^{ij}$, it will not influence bulk, experimentally measurable transport quantities which can be computed from $L^{ij}$, such as the ionic conductivity. We also note that $\mathbf{v}_i$ is technically an average over all instances of species $i$ in the system (e.g. over every cation of a given type in the electrolyte): $\mathbf{v}_i = \frac{\sum_\alpha^{N_i} v_{i,\alpha}(t)}{N_i}$. Thus, we rewrite our original expression for $L^{ij}$ as follows:

$L^{ij} = \frac{V}{3 k_B T} \int_0^\infty \big<\frac{N_i N_j}{V^2}\frac{\sum_\alpha^{N_i} v_{i,\alpha}(t)}{N_i} \cdot \frac{\sum_\beta^{N_j} v_{j,\beta}(0)}{N_j}\big>$

$L^{ij} = \frac{1}{3 k_B T V} \int_0^\infty \big<\sum_\alpha v_{i,\alpha}(t) \cdot \sum_\beta v_{j,\beta}(0)\big>$

For computational ease, this final form of the equation is the one evaluated in this code.

### Preparation of input files

This example is for a simulation of 0.1 M LiCl in dimethyl sulfoxide. In the working directory, we need: 

1. the lij_analysis.py analysis script

2. LAMMPS dump files for the ion velocities (e.g. v_anion.out and v_cation.out). 
        
Example portion of LAMMPS input script to generate dump files:
            # Calculate velocities of anions
            group anion type 2  # creat group containing all anions
            dump anion_dump anion custom 1 v_anion.out id vx vy vz  # dump velocities every time step

            # Calculate velocities of cations
            group cation type 3
            dump cation_dump cation custom 1 v_cation.out id vx vy vz
            
For the purposes of uploading to this repository, the example velocity output files were split into smaller files. These are combined back into v_anion.out and v_cation.out below.

```python
from lij_analysis import TransportCoefficients
import numpy as np
import matplotlib.pyplot as plt
```

### Generate times array
For this example simulation, velocity was dumped every 2 fs for a total of 5 ns.

```python
dt = 2  # 2 fs between velocity data collection
t_total = 5e6  # 5 ns (5e6 fs) total data collection time
times = np.arange(0,t_total + dt, dt)
```

### Input box volume

```python
V = 118259.53  # cubic Angstroms, obtained from LAMMPS log file
```

### Prepare velocity files
Concatenate velocity output files together. These were split such that smaller files could be uploaded to the repository.

```python
! cat example_velocity_output/v_anion_* > v_anion.out
! cat example_velocity_output/v_cation_* > v_cation.out
```

## Compute transport coefficients

```python
test_analysis = TransportCoefficients('v_cation.out','v_anion.out', V, times)
```

```python
acf_plusplus, acf_plusminus, acf_minusminus, L_plusplus, L_plusminus, L_minusminus = test_analysis.compute_all()
```

### Plot x-component of correlation function

```python
plt.plot(times, acf_plusplus[:,0], label = '+ +')
plt.plot(times, acf_plusminus[:,0], label = '+ -')
plt.plot(times, acf_minusminus[:,0], label = '- -')
plt.xlim(0,1000)
plt.ylim(-0.0002, 0.0002)
plt.legend()
plt.ylabel('Correlation Function')
plt.xlabel('Time (ps)')
plt.show()
```

### Plot transport coefficients

```python
plt.plot(times[:-1], L_plusplus, label = '+ +')
plt.plot(times[:-1], L_plusminus, label = '+ -')
plt.plot(times[:-1], L_minusminus, label = '- -')
plt.legend()
plt.xlim(0,30000)
plt.ylim(-0.5e37,0.3e38)
plt.xlabel('Time (fs)')
plt.ylabel('$L^{ij}$ / (J s m)$^{-1}$')
plt.show()
```

### Use transport coefficients to compute conductivity
$\kappa = \sum_i \sum_j L^{ij} z_i z_j F^2$

```python
e2c = 1.60217662e-19  # convert electronic charge to Coulombs
conductivity = (L_minusminus + L_plusplus - 2*L_plusminus)*10*(e2c)**2 # mS/cm
```

```python
plt.plot(times[:-1], conductivity)
plt.xlim(0,30000)
plt.ylim(0, 10)
plt.xlabel('Time (fs)')
plt.ylabel('Conductivity / ms cm$^{-1}$')
plt.show()
```

```python
# Choosing cutoff time to be, for example, 20000 ps:
print "Final conductivity: " + str((conductivity)[20000/2]) + " mS/cm"
```
